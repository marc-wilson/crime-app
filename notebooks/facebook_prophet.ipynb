```python
import pandas as pd
from fbprophet import Prophet
```

```python
crime_df = pd.read_csv('data/crimes_by_hour.csv')
prophet_df = crime_df.copy()
prophet_df['ds'] = prophet_df['Date']
prophet_df['y'] = prophet_df['Count']
prophet_df = prophet_df.drop(['Date', 'Count'], axis=1)
prophet_df = prophet_df.drop(prophet_df.columns[prophet_df.columns.str.contains('unnamed', case=False)], axis=1)
prophet_df.head()
```

```python
m = Prophet()
m.fit(prophet_df)
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```
